# Introduction
The relationship between wages and workers has been a crucial topic in labor economics for decades. Having a quantitative relationship between these two can allow for workers to understand the value of their work and abilities and lead to more satisfactory job choices, and firms can create more accurate policies and compensation structures to reflect the benefit gained from the workers they have employed and want to employ in the future. 

Several academic papers have discussed different approaches meauring the relationship between worker compensation and their tenure. Here we review some literature on similar topics, and the results they obtained.

- "Wages, Experience and Seniority" by Christian Dustmann and Coastas Meghir report findings on how a workers' compensation changes depending on certain factors. Specifically the report mentioned creates a function based on explanatory variables found in their dataset, and gives a result of a numerical growth rate for each year of employment. (Dustmann,Meghir, 2005).

- "Do Wages Rise with Job Seniority?" by Joseph Altonji and Robert Shakotko discusses the topic by using a dataset of workers in the 60's which include their ages and disabilities along with their tenure and salaries, and creates an equation to try and capture the effect the mentioned explanatory variables have on the workers wages. The paper reports that the effect of tenure on wages is small, and that the workers gain wage growth from their overall experience. (Altonji, Shakotko, 1985).

- "Job Duration, Seniority, and Earnings." by Katharine Abraham and Henry Farber uses the same dataset as Altonji and Shakotko, but instead tries to determine how much the effect of reported salary growth actually comes from experience. The paper finds once again that experience doesn't actualy have much of an effect on salary growth, which suggests the hypothesis that wages and tenure are not related. (Abraham, Farber, 1987). Continuing off this idea, we look at a paper that talks about another factor that explains how worker's salary is determined, the company said worker is employed.

- "The Wage Policy of a Firm" by George Baker, Michael Gibbs, and Bengt Holmstrom uses another imperical dataset from workers in Management. The paper uses a model that tries to attribute wage growth to the worker's comparative advantage in their workplace. However, the paper finds that the learning model used doesn't explain growth much, and that the method in which firms compensate their workers come from a predetermined salary increase that workers must compete for. (Baker, 1994) 

The findings in the literature suggest that the relationship between worker salary and experience is much more complicated than expected, and that their isn't much to report on in terms of new findings. Now, we obtain relevent statistics of the cleaned data. First, it will be helpful to the average salaries for each variable to see general trends. Below are general purpose functions that can be used on specified columns to find the relevent statistics. The first function finds frequency / popularity of a given column, and just returns a series. The second function requires both the salary column to find an average of, and other data columns that are transformed into a list. The list variable list_of_x will be manually made before each use of the function. Both will be called upon heavily for data collection and interpretation.

```python
def freq_name(data_frame, column_name):
    """Given a dataframe and a column name, return a Series of each company along with the number of times they appear.
    """
    if column_name not in data_frame.columns:
        print('Invalid Column')
        return None

    entry_counts = data_frame[column_name].value_counts()
    return entry_counts


def average_y_in_x(df, column1, column2):
    """Returns a dictonary that has keys from list_of_x, and values be average entries of column1 that have the key value in column 2.
    """
    if column1 not in df.columns:
        print('Invalid Column')
        return None
    
    if type(list(set(dfnew[column1].tolist()))[0]) == str:
        list_of_x = list(set(dfnew[column1].tolist()))
    
    elif type(list(set(dfnew[column1].tolist()))[0]) == float:
        _list = list(set(dfnew[column1].tolist()))
        list_of_x = list(set([math.floor(x) for x in _list]))
        
    d = {}
    for category in list_of_x:
        filtered_df = df[df[column1] == category] 
        d[category] = filtered_df[column2].mean().round(0)
    return d
```

Using these functions, we then create individual DataFrames for each important category to salary. With respect to coding, the .iloc command will be used to obtain years to salary data for 3 year periods, to show periodic growth rather than year by year. We also want to include 8 rows, as thats the total number of job categories, and allows all the other DataFrames to show their top entries and the lower ones aswell.

```python
top_companies = freq_name(dfnew, 'employer_name')
top_countries = freq_name(dfnew, 'location_country')
top_categories = freq_name(dfnew, 'job_title_category')
average_sal_per_category = average_y_in_x(dfnew,'job_title_category','annual_base_pay')
average_sal_per_year = average_y_in_x(dfnew,'total_experience_years','annual_base_pay')
average_sal_per_3year_df = pd.DataFrame.from_dict(average_sal_per_year, orient = 'index').iloc[::3]
average_sal_per_category_df = pd.DataFrame.from_dict(average_sal_per_category, orient = 'index')

df1 = top_companies.head(4).to_frame()
df1.columns = ['# of workers']
df2 = top_countries.head(4).to_frame()
df2.columns = ['# of workers']
df3 = top_categories.head(8).to_frame()
df3.columns = ['# of workers']
df4 = average_sal_per_category_df.head(8)
df4.columns = ['Average Salary']
df5 = average_sal_per_3year_df.head(8)
df5.columns = ['Average Salary Every 3 Years']
df3and4 = pd.merge(df3, df4, left_index = True, right_index = True)

frequency_dict = {}
for title in dfnew['job_title'].tolist():
    if title in frequency_dict:
        frequency_dict[title] += 1
    else:
        frequency_dict[title] = 1
sorted_title_dict = dict(sorted(frequency_dict.items(), key=lambda x: x[1], reverse=True))
df6 = pd.DataFrame.from_dict(sorted_title_dict, orient='index', columns=['# of workers'])
df6.index.name = 'Title'
df6 = df6.head(3)

html_str = '<table><tr><td>{}</td><td>{}</td><td>{}</td><td>{}</td><td>{}</tr></table>'.format(df1.to_html(), df2.to_html(), df3and4.to_html(), df5.to_html(),df6.to_html())
display(HTML(html_str))
```

```python
def average_growth_rate(numbers):
    growth_rates = []
    for i in range(1, len(numbers)):
        growth_rate = ((numbers[i] - numbers[i-1]) / numbers[i-1])
        growth_rates.append(growth_rate)

    average_rate = sum(growth_rates) / len(growth_rates)
    return round(average_rate, 4)

def find_average_salary_growth(df, location, category):

    df_location= df[(df['location_state'] == location) & (df['job_title_category'] == category)]

    lst_of_years_worked = []
    for year in df_location['employer_experience_years']:
        if year not in lst_of_years_worked:
            lst_of_years_worked.append(year)   
    lst_of_years_worked.sort()
    lst_of_years_worked

    average_sal_lst = []
    for year in lst_of_years_worked:
        dfnew_year = df_location[df_location["employer_experience_years"] == year]
        average_sal_peryear = dfnew_year['annual_base_pay'].mean()
        average_sal_lst.append(round(average_sal_peryear,2))

    return average_growth_rate(average_sal_lst)

def find_average_salary_lst(df, location, category):

    df_location= df[(df['location_state'] == location) & (df['job_title_category'] == category)]

    lst_of_years_worked = []
    for year in df_location['employer_experience_years']:
        if year not in lst_of_years_worked:
            lst_of_years_worked.append(year)   
    lst_of_years_worked.sort()
    lst_of_years_worked

    average_sal_lst = []
    for year in lst_of_years_worked:
        dfnew_year = df_location[df_location["employer_experience_years"] == year]
        average_sal_peryear = dfnew_year['annual_base_pay'].mean()
        average_sal_lst.append(round(average_sal_peryear,2))
    return [lst_of_years_worked, average_sal_lst]
```

## Potential Data to Scrape


Data that will complement the existing dataset of tech salaries will be information that can help further distinguish the difference between nominal and real salary growth rates. An immediate issue when trying to observe a workers true salary growth to their experience is whether or not they are gaining any increase in salary. Comparing different worker's salary growth is only significant if their unqiue costs of living are held constant. For example, if the cost of living goes up more than the annual salary growth rate, the workers are really making a negative growth rate over all. Therefore, finding some measure of cost of living in the different locations will be economically important. 

The ideal place to find cost of living will be a website that has historical data and several different categories such as average rent prices and average transporation costs. Furthermore, it would be more accurate if there was a distinsion in the website between rural and urbern parts of the location's costs. Lastly, the website should be universal for different cities, since if it wasn't, the only way to collect data would be refering to a differnt site for each location, which could easily lead to inaccuracies and defeats the purpose of adding this new data. 

Fortunately, the website "https://numbeo.com/" satisfies all of the above mentioned necessities, and is easily available to scrape. Using the scraped data, we can merge the cost of living growth rates to each row entry given their geographical location, and find an adjusted annual salary with a cost of living index relative to a chosen baseline cost.

## Potential Challenges

In the "numbeo" website, there aren't really any difficulties in scraping the information. One possible concern is the accuracy of the reported data, since it uses user inputed data, however since the original tech salaries dataset was also surveyed and user input, the accuracy concern isn't a big concern. ## Scraping Data from A Website

```python
def find_average_living_cost(City):
    years = [1,2,3,4,5,6,7,8,9]
    years.reverse()
    lst = {}
    # We find the historical growth rate beginning from 2013 up until 2021, to then find a more accurate growth rate.
    for year in years:
        formated_name = City.replace(' ', '-')
        web_url = 'https://www.numbeo.com/cost-of-living/city-history/in/' + formated_name
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.190 Safari/537.36',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
            'Accept-Encoding': 'gzip, deflate, br',
            'Accept-Language': 'en-US,en;q=0.9',
            'Connection': 'keep-alive',
        }

        response = requests.get(web_url, headers = headers)

        soup_object = BeautifulSoup(response.content)
        #With the website initialized and a soup object created, we can now start examining the different tables on the website.
        #The format of Numbeo has the Resturanet price in the 1st table, Rental costs in the 6th table, Transporation pass in the 10th
        # and the utilities price in the 12th table.
        
        
        data_table = soup_object.find_all("table")[6]
        all_values = data_table.find_all('tr')
        split_text = all_values[year].text #index is row
        rent_1bdrm = split_text.split()[1]

        data_table = soup_object.find_all("table")[12]
        all_values = data_table.find_all('tr')
        split_text = all_values[year].text #index is row
        utlities_price = split_text.split()[1]

        data_table = soup_object.find_all("table")[1]
        all_values = data_table.find_all('tr')
        split_text = all_values[year].text #index is row
        resturant_price = split_text.split()[1]

        data_table = soup_object.find_all("table")[10]
        all_values = data_table.find_all('tr')
        split_text = all_values[year].text #index is row
        pass_price = split_text.split()[1]
        
        
        #Finally, we sum each monthly cost and append the value to a list, which we wil then take the average growth rate of. 
        #Note we assume the number of meals a workers get at a resturant is 12. While this is arbitrary, it doesn't affect the cost of living much anyways.
        monthly_cost = float(rent_1bdrm) + float(utlities_price) + 12 * float(resturant_price) + float(pass_price)
        lst[2024 - year ] = round(monthly_cost,2)                                                                                 
    return lst

#The following function takes a list of numbers and just finds the average growth rate of the entrees, which is a more accurate
# way of finding growth rates across time than finding the secant line of the inital time period and final time period
def average_growth_rate(numbers):
    growth_rates = []
    for i in range(1, len(numbers)):
        growth_rate = ((numbers[i] - numbers[i-1]) / numbers[i-1])
        growth_rates.append(growth_rate)

    average_rate = sum(growth_rates) / len(growth_rates)
    return round(average_rate, 4)

#This function does the same thing as the webscrapping, but with the dataset across different job categories in location=
def find_average_salary_growth(df, location, category):

    df_location= df[(df['location_state'] == location) & (df['job_title_category'] == category)]
    #We find the number of unique tenure lengths in a given state, then find the average salary growth weighted for the frequency of each tenure length
    lst_of_years_worked = []
    for year in df_location['employer_experience_years']:
        if year not in lst_of_years_worked:
            lst_of_years_worked.append(year)   
    lst_of_years_worked.sort()
    lst_of_years_worked

    average_sal_lst = []
    for year in lst_of_years_worked:
        dfnew_year = df_location[df_location["employer_experience_years"] == year]
        average_sal_peryear = dfnew_year['annual_base_pay'].mean()
        average_sal_lst.append(round(average_sal_peryear,2))

    return average_growth_rate(average_sal_lst)
```

```python
def yahoo_lookup(symbol):
    max_attempts = 4
    attempts = 0
    while attempts < max_attempts:
        year = str(2016+attempts)
        start_date = year + '-01-01'
        end_date = year + '-12-31'
        try:
            historical_data = si.get_data(symbol, start_date, end_date)
            close_stock_price = round(historical_data['close'].tolist()[0], 2)
            volume = round(historical_data['volume'].tolist()[0], 2)
        except Exception as e:
            attempts += 1
            close_stock_price = float('nan')
            volume = float('nan')             
        else:
            break        
    return [close_stock_price, volume]
```

### Regression Explanation

A linear regression is an equation that has an output number (Y) on the left, and a series of terms of different variables (X) that try to explain Y. The coefficients infront of each X variable represents the effect of the X variable on changing Y numerically. The general form for a linear regression that has multiple X variables is as follows;

$$
Y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_k x_k + u
$$

- $\beta_0 $: The intercept term

- $\beta_i $: The i-th coefficient for the i-th explanatory variable, which represents the effect on Y by increasing $x_i$ by 1.

- $u$: The error term, which contains all unexplained impact on Y that are not in the X variables.

This is just the general linear form however, and there are many other ways to express a regression in both a linear and non-linear way, which uses binary, exponentional, log, and interactive x-variables. Some of these will be used in the regression analysis later on this section. Running the regressions is now finding predicted values for Y and the $\beta_i$, using the dataset that contains both the Y and X variables. The predicted values are represented by $\hat{Y}$ and $\hat{\beta_i}$ respectively. The method for finding the best predicted values is by using OLS.

OLS stands for Ordinary Least Squares, which is the method that minimizes the differnence between the real Y values, and the predicted Y values, finding the $\beta_i$ that minimize the sum of squares across all row entrees 1, ... , n. 

$$
(\hat{\beta_0}, \hat{\beta_1}, \cdots, \hat{\beta_k}) = \underset{\beta_0,\beta_1,\cdots,\beta_k}{\operatorname{argmax}} \dfrac{1}{n}\sum_{i=1}^{n} ( y_i - b_0 - b_1 x_{i1} - \cdots -  b_k x_{ik})^2
$$

Once these predicited values are found, they can be examined to find the effect a given coefficient has on the output variable Y. All of this information was found on in â€œIntroductory econometrics: a modern approachâ€, by Wooldridge.

The chosen Y was Annual Salary, and the main X was Experience. Under economic theory, the relationship should be linear as there should be relatively constant changes in salary each year. The gain in salary for an inexperienced worker might be less than the gain a more experienced worker will get, but at the same time, the more experienced worker isn't taking on a marginally high extra workload to compensate for the higher salary, so returns should even out. If the quesiton was a worker's market value with experience, then I would believe the relationship to be nonlinear and probably quadratic with age, but at the same job position and title the relationship should be linear.

To remove the high amount of missings in some of the important columns, we can implement Data Imputation technqiues to fill in the missing values. Literature on the topic suggests to only use Median or Mean data imputation when only a few entrees are missing, but for example, the bonuses columns are missing almost 20% of their entrees each. (Zhang, 2016). The same paper discusses using regressions to preduce "smarter" values, but talks about the issue with this method. By the nature of imputating with regression, the imputated data will cause the bonuses columns to be highly correlated with the other columns. However, this shouldn't be a big problem as there will still be low enough correlation in the data to safely run an accurate regression.

```python
import statsmodels.api as sm
X=['const','employer_experience_years','prior_experience','job_category_binary_Software', 'job_category_binary_Management',
                 'job_category_binary_Engineering','state_binary_CA','state_binary_NY', 'state_binary_WA',
                 'job_title_binary_SE','job_title_binary_SSE','job_title_binary_SD']                 


#reg_abonus = statsmodels.formula.api.OLS(merged_df['annual_bonus'], merged_df[X], missing='drop').fit()
reg_abonus = sm.OLS(merged_df['annual_bonus'], merged_df[X], missing='drop').fit()
reg_sbonus = sm.OLS(merged_df['signing_bonus'], merged_df[X], missing='drop').fit()

info_dict={'R-squared' : lambda x: f"{x.rsquared:.2f}", 
           'No. observations' : lambda x: f"{int(x.nobs):d}"}

data_imputation_table = summary_col(results=[reg_abonus,reg_sbonus], float_format='%0.2f', stars = True, model_names=['Model 1',
                                    'Model 2'], info_dict=info_dict,
                                    regressor_order = X)
data_imputation_table.add_title('Data Imputation Regression')
#print(data_imputation_table)
```

```python
def create_ols_eq(reg):
    ols = ''
    parameters = reg.params
    parameters_dict = parameters.to_dict()
    parameters_dict
    for var in parameters_dict:
        beta = round(parameters_dict[var],3)
        
        if var == 'const':
            ols = str(beta)
        elif '-' in str(beta):
            ols = ols + " + " + str(beta) + var
        else:
            ols = ols + ' + ' + str(beta) + var
    return ols
def contains_text(input_string):
    for char in input_string:
        if char.isalpha():
            return True
    return False
def split_string(input_string):
    numbers_part = ''
    text_part = ''
    for char in input_string:
        if char.isdigit() or char == '.' or char == '-':
            numbers_part += char
        else:
            text_part += char
    return [float(numbers_part), text_part]
def equation_to_dict(equation_string):
    equation = create_ols_eq(equation_string)
    terms = equation.split(' + ')
    coefficient_dict = {}
    for i in range(0,len(terms)):
        #print(terms[i])
        if contains_text(terms[i]):
            #print(split_string(terms[i]))
            coefficient_dict[split_string(terms[i])[1]] = split_string(terms[i])[0]
        else:
            coefficient_dict['const'] = float(terms[i])    
    return coefficient_dict
```

```python
regression_dict_abonus = equation_to_dict(reg_abonus)
missing_rows = merged_df['annual_bonus'].isnull()
for index, row in  merged_df[missing_rows].iterrows():
    predicted_value = regression_dict_abonus['const']
    for column in list(regression_dict_abonus.keys())[1:]:
        predicted_value += row[column] * regression_dict_abonus[column]
    merged_df.loc[index, 'annual_bonus'] = predicted_value
    
    
regression_dict_sbonus = equation_to_dict(reg_sbonus)
missing_rows = merged_df['signing_bonus'].isnull()
for index, row in  merged_df[missing_rows].iterrows():
    predicted_value = regression_dict_sbonus['const']
    for column in list(regression_dict_sbonus.keys())[1:]:
        predicted_value += row[column] * regression_dict_sbonus[column]
    merged_df.loc[index, 'signing_bonus'] = predicted_value
```

### Objective function for Regression Tree

Since the prefered specficiation is regression 8, we can reuse the explanatory variables and fit them to the decision tree model. In this case, a new dataframe will be created that copies over the columns used in regression 8, and the output value will be the annual salary, undergoing a log transformation to adjust for size.

The following equation is the objective function that iteratively creates different branches of X values, all while trying to minimize the mean squared error, the value that shows the differenc in average deviation to actual target deviation. The equation is run on each new region created by the previous split, which is why it minimizes the sum of the squared differences, where $y_i$ is the real salary growth value at that rectangle, and $\hat{y}_{R1}$ is the predicted salary growth value in that rectangle.
  
$$
\min_{j,s} \left[  \sum_{i: x_{i,j} \leq s, x_i \in R1}
    (y_i - \hat{y}_{R1})^2 +  \sum_{i: x_{i,j} > s, x_i \in R2}
    (y_i - \hat{y}_{R2})^2 \right]
$$

After a specified number of regions are formed, which will be noted as the depth of the tree, the function will stop looping, and then a step called 'pruning' will take place, which essentially removes branches that do not drastically change the predictive power of the model. The reason this is done is to prevent over-fitting, which occurs when too many branches are made and not enough distinction between splits are considered with each new split. The depth, number of regions, and pruning amount are all considered to be regularization parameters, which can be adjusted for accuracy.
